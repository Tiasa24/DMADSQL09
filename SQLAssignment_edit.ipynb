```python
import numpy as np
import pandas as pd
from faker import Faker
fake = Faker()                  #initializing
```

```python
Faker.seed(20) 
```

### 1. Creating columns for Member Table

```python
n = 1000

#Creating fake 1000 names
names = [fake.name() for _ in range(n)]

#Creating membership ids
membership_id = ["MEM01", "MEM02", "MEM03", "MEM04", "MEM05", "MEM06", "MEM07", "MEM08"]
probabilities = [0.25, 0.10, 0.10, 0.10, 0.20, 0.10, 0.10, 0.05] 
membership_ids= np.random.choice(membership_id, size=n, p=probabilities) 

#Creating Member ids
member_ids = [f"MEMB{num}" for num in np.random.choice(range(1000, 9999), size=n, replace=False)]

#Creating Phone number with missing value 
phone = [f"{np.random.randint(600, 999)}-****-{np.random.randint(100, 999)}" for _ in range(n)]

phone = np.where(np.random.rand(n) < 0.2, None, phone)

#Creating Email id 
email = [f"{name.split()[0].lower()}{name.split()[1].lower() if len(name.split()) > 1 else ''}
        {np.random.randint(100, 999)}@gymmail.com" for name in names]

email = np.where(np.random.rand(n) < 0.12, None, email)
#Creating Joining date
join_year = np.random.randint(2023, 2025, n)  # 2023-2024
join_date = [
    f'{join_year[i]}-01-{str(np.random.randint(1, 31)).zfill(2)}' 
    if np.random.random() < 0.6 else 
    f'{join_year[i]}-{str(np.random.randint(1, 13)).zfill(2)}-' 
    f'{str(np.random.randint(1, 29)).zfill(2)}'
    for i in range(n)
]
join_date = np.where(np.random.rand(n) < 0.1, None, join_date)

#Creating Address
postcode = [f'AL{str(i).zfill(2)}' for i in range(1, 11)]
postcode_data = np.random.choice(postcode, n)

postcode_data = np.where(np.random.rand(n) < 0.15, np.nan, postcode_data)

# Introduce duplicate names 
duplicates = int(0.003 * n)
duplicate_indices = np.random.choice(n, duplicates, replace=False)
for i in duplicate_indices:
    names[i] = np.random.choice(names)  # Replace with a random existing name

# Introduce duplicate phone numbers 
duplicate_indices = np.random.choice(n, duplicates, replace=False)
for i in duplicate_indices:
    phone[i] = np.random.choice([p for p in phone if p is not None]) 
    
Table1 = pd.DataFrame({
    'Name' : names,
    'MemberId' : member_ids, 
    'MembershipId': membership_ids,
    'Phone' : phone,
    'EmailId' : email,
    'Address': postcode_data,
    'Joining_Data': join_date
})



Table1.to_csv('Member.csv', index = False)
Table1.head()
```

### Creating columns for Membership Type Table

```python
membership_id = ["MEM01", "MEM02", "MEM03", "MEM04", "MEM05", "MEM06", "MEM07", "MEM08"]

membership = ['Individual','Teen-Junior', 'Joint', 'Group', 'Family','Corporate Individual','Corporate Group (up to 4)','Students']

access_levels = ["Basic", "Standard","Premium", "VIP"]
access = np.random.choice(["Basic", "Standard","Premium", "VIP"], size=len(membership_id), p=(0.5,0.3,0.1,0.1))

monthly_price = np.round(np.random.uniform(30, 150, 8), 2)
yearly_price = np.round(np.random.uniform(300, 1800 ,8),2)

Table2 = pd.DataFrame({ 
    'MembershipId' : membership_id,
    'Membership_Type' : membership,
    'Monthly_Fee' : monthly_price,
    'Yearly_Fee' : yearly_price,
    'Access_level' : access
                                      })

Table2.to_csv('Membership_Type.csv', index =False)
```

### Creating columns for Payment Table

```python
import string
n = 1000

payment_ids = set()
while len(payment_ids) < n:
    new_id = f"{np.random.randint(1000, 9999)}{np.random.choice(list(string.ascii_uppercase))}{np.random.choice(list(string.ascii_uppercase))}"
    payment_ids.add(new_id)

payment_ids = list(payment_ids)
payment_methods =np.random.choice(['Credit Card', 'Cash', 'Bank Transfer'], n, p = [0.45,0.25,0.30])

amount_paid = np.where(
    payment_methods == 'Credit Card', 
    np.round(np.random.uniform(100, 200, n), 2),  # Middle range for Credit Card
    np.where(
        payment_methods == 'Cash', 
        np.round(np.random.uniform(30, 100, n), 2),  # Low range for Cash
        np.round(np.random.uniform(200, 1000, n), 2)  # High range for Bank Transfer
    )
)
amount_paid = np.where(np.random.rand(n) < 0.23, np.nan, amount_paid)
Table3 = pd.DataFrame({
    'Payment_ids' : payment_ids,
    'MemberId' : member_ids,
    'Payment_Method' : payment_methods,
    'Amount_Paid': amount_paid
})

Table3.to_csv('Payment.csv', index=False)
Table3.head(10)
```

### Creating columns for Class Table
